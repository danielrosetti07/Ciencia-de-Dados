```python
# 18.Elabore uma função chamada calcula_desconto que receba dois parâmetros: o valor de um produto e a porcentagem de desconto. A função deve retornar o valor final do produto com o desconto aplicado. Teste a função e imprima o resultado.
def calcula_desconto(valor, porcentagem):
    desconto = valor * (porcentagem/100)
    valor_final = valor - desconto
    return valor_final

preco = 100
desconto = 20

resultado = calcula_desconto(preco, desconto)
print(f"Valor final com desconto: {resultado}")
```

```python
# 19.Elabore uma função chamada classifica_nota que receba uma nota de 0 a 10. A função deve retornar "Aprovado" se a nota for maior ou igual a 7, "Recuperação" se estiver entre 5 e 6.9, e "Reprovado" se for menor que 5. Chame a função passando uma nota e imprima o retorno.
def classifica_nota(nota):
  if nota >= 7:
    return "Aprovado!"
  elif nota >= 5:
    return "Recuperação"
  else:
    return "Reprovado"

resultado = classifica_nota(6.5)
print(f"Resultado: {resultado}")
```

```python
# 20.Elabore um código que receba uma lista de números inteiros mistos (positivos, negativos e zeros). Utilize um laço for com estruturas condicionais if/elif/else para contabilizar quantos números são positivos, quantos são negativos e quantos são exatamente zero. Ao final, imprima esse resumo.

numeros = [10, 8, 0, 4, 5, 1, 7, 0, -1, -8, -3]

positivos = 0
negativos = 0
zeros = 0

for n in numeros:
    if n > 0:
       positivos += 1
    elif n < 0:
       negativos += 1
    else:
       zeros += 1

print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Zeros: {zeros}")
```
